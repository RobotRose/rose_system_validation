```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
raw = pd.read_csv("/home/osch/connection_debug/wlan.csv")
log = pd.DataFrame()
```

```python
raw.index = raw["Unnamed: 0"]
```

```python
# raw.tail()
```

```python
APs = raw["Access Point"].unique()
filtered_APs = [ap for ap in APs if ":" in ap]

map_APs = dict(zip(filtered_APs, range(100)))
print map_APs
router_descriptions = {' 38:2C:4A:66:97:B0': "R2 on pavilion", ' 10:C3:7B:DE:58:58': "Router in office", ' 38:2C:4A:66:E9:40': "R1 on hallway"}
```

```python
def normalize(df):
#     return column - column.mean() / (column.max() - column)
    return df/df.loc[df.abs().idxmax()].astype(np.float64)
```

```python
map_onoff = {"off":0, "on":1}

def clamp(n, minn, maxn):
    return max(min(maxn, n), minn)

def onoff_to_01(val):
    return map_onoff.get(val, np.nan)

def accespoint_to_index(val):
    return map_APs.get(val, -1)

def fraction_to_float(val):
    try:
        parts = val.split("/")
        return float(parts[0]) / float(parts[1])
    except:
        return np.nan
    
def dBm_to_float(val):
    try:
        val = val.strip("dBm").strip()
        return float(val)
    except:
        return np.nan
    
def str_to_int(val):
    try:
        val = val.strip()
        return int(val)
    except:
        return np.nan

boolean_cols = ["Power Management", "Fragment thr", "RTS thr"]
numeric_cols = ["Tx excessive retries", "10.8.0.1_ttl", "10.8.0.1_time", "10.8.0.5_ttl", "10.8.0.5_time"]
fraction_cols = ["Link Quality"]
dBm_cols = ["Signal level"]

for col in boolean_cols:
    log[col] = raw[col].apply(onoff_to_01)
for col in fraction_cols:
    log[col] = raw[col].apply(fraction_to_float)
for col in dBm_cols:
    log[col] = raw[col].apply(dBm_to_float)
for col in numeric_cols:
    log[col] = raw[col].apply(str_to_int)
    
for col in boolean_cols+fraction_cols+dBm_cols+numeric_cols:
    log[col] = normalize(log[col])
    
for col in [c for c in raw.columns if not c in log.columns]:
    log[col] = raw[col]
    
log["Access Point"] = raw["Access Point"].apply(accespoint_to_index)
log["AP Switch"] = log["Access Point"].diff().abs()#/ * 2

del log["Power Management"]
del log["Fragment thr"]
del log["RTS thr"]
del log["10.8.0.1_ttl"]
del log["10.8.0.1_time"]
del log["10.8.0.5_ttl"]

# log.tail()
```

```python
log.plot(rot=90) # , legend=False
```

```python
corr = log.corr()#.abs()
plt.yticks(np.arange(0.0, len(log.columns), 1), corr.columns)
plt.xticks(np.arange(0.0, len(log.columns), 1), corr.columns, rotation=90)
plt.imshow(corr, interpolation='none', cmap=plt.cm.Blues) #See http://matplotlib.org/examples/color/colormaps_reference.html
corr
```
